```python
# red, 별모양, dash선으로 sin 그리기
# green, 동그라미, 실선으로 cos 그리기

theta = np.linspace(0, 2 * np.pi, 20)
plt.plot(theta, np.sin(theta), 'r*--', markersize = 10, label = "sin")
# plt.figure() # 새창 열기, 이거 없으면 겹쳐 그림
plt.plot(theta, np.cos(theta), 'go-', markersize = 10, label = "cos")
plt.xlabel(r"$theta$") # r은 raw
plt.ylabel("value")
plt.title("sin and cos", color = "b")
plt.legend() # label 뜨게하기
plt.grid() # 격자 생성
# plt.xlim([0,np.pi])
# plt.ylim([-1.5, 1.5])
plt.axis([0, np.pi, -1.5, 1.5]) # 위 13, 14를 합쳐놓은 것과 동일
```

```python
x = range(5)
y1 = [i for i in range(1, 6)]
y2 = [i+100 for i in range(1, 6)]

plt.figure()
plt.plot(x, y1)
plt.plot(x, y2)

plt.figure()
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
p1 = ax1.plot(x, y1, 'b')
p2 = ax2.plot(x, y2, 'r')
ax1.set_ylim([-10, 10]) # ax1.ylim([]) 사용 못함
ax1.set_xlabel("x")
ax1.set_ylabel("y1")
ax2.set_ylabel("y2")
ax1.set_title("y1 and y2")
ax1.grid()
plt.legend(p1 + p2, ["y1", "y2"], loc = "best") # loc로 위치 지정
```

2개 변수 함수, 등고선으로 보기, 3D로 보기

```python
a = np.linspace(-1, 1, 100)
b = np.linspace(-1, 1, 100)
A, B = np.meshgrid(a, b)

L = A**2 + B**2

plt.contour(a, b, L, 20) # 20 -> 등고선 수
plt.xlabel("a"); plt.ylabel("b"); plt.grid()
```

```python
plt.figure(figsize = [10, 9]) # [가로, 세로]
ax = plt.axes(projection = "3d")
ax.plot_surface(A, B, L)

plt.figure(figsize = [10, 9])
ax = plt.axes(projection = "3d")
ax.plot_wireframe(A, B, L)
```

```python
figure = plt.figure(figsize = [10, 9]) # [가로, 세로]
ax = figure.add_subplot(projection = "3d") # figure.gca 사용시 에러로 인해 코드 변경
ax.view_init(elev = 10, azim = 45)
ax.plot_surface(A, B, L)

figure = plt.figure(figsize = [10, 9])
ax = figure.add_subplot(projection = "3d")
ax.view_init(elev = 10, azim = 45)
ax.plot_wireframe(A, B, L)
```

```python
from matplotlib.colors import LightSource
plt.figure(figsize = [10, 9])
ax = plt.axes(projection = "3d")
ls = LightSource(altdeg = 10, azdeg = 45)
rgb = ls.shade(L, plt.cm.RdYlBu)
# ax.plot_surface(A, B, L, facecolors = rgb)
# 4-6라인 대신 8라인으로 표현 가능
# ax.plot_surface(A, B, L, cmap = 'viridis') # viridis, plasma, inferno, magma, cividis

ax.plot_surface(A, B, L, facecolors = rgb, rstride = 1, cstride = 1, linewidth = 0, antialiased = False, shade = False)
```

```python
plt.figure(figsize = [10, 9])
ax = plt.axes(projection = "3d")
ls = LightSource(altdeg = 10, azdeg = 45)
rgb = ls.shade(L, cmap = plt.cm.gist_earth, vert_exag = 0.1, blend_mode = 'soft')
ax.plot_surface(A, B, L, rstride = 1, cstride = 1, facecolors = rgb, linewidth = 0, antialiased = False, shade = False)
```

```python
a = np.linspace(-1, 1, 10)
b = np.linspace(-1, 1, 10)
A, B = np.meshgrid(a, b)
L = A**2 + B**2

plt.figure(figsize = [10, 9])
ax = plt.axes(projection = "3d" )
ax.plot(A.reshape(-1, 1).squeeze(), B.reshape(-1, 1).squeeze(), L.reshape(-1, 1).squeeze(), "ro")

plt.figure(figsize = [10, 9])
ax = plt.axes(projection = "3d" )
ax.scatter(A, B, L, s = 100, c = "g", marker = "*", edgecolor = "r", alpha = 1) # s 사이즈, alpha 투명도
```

```python
plt.figure(figsize = [10, 9])
ax = plt.axes(projection = "3d")
ax.plot_surface(A, B, L, alpha = 0.3, cmap = "viridis")
ax.scatter(A, B, L, s = 100, c = "r", marker = "*")
```

plotly

```python
import plotly.graph_objects as go

a = np.linspace(-2, 2, 100)
b = np.linspace(-2, 2, 100)
A, B = np.meshgrid(a, b)
L = np.cos(A**2.4) + np.sin(B**2.4)

fig = go.Figure(data = [go.Surface(x = a, y = b, z = L, colorscale = "viridis", opacity = 0.5)])
fig.update_traces(contours_z = dict(show = True, usecolormap = True, highlightcolor = "limegreen", project_z = True)) # project_z 컨투어
fig.update_layout(title = "cos(A^2.4) + sin(B^2.4)", width = 700, height = 600)
```

```python
import plotly.graph_objects as go

a = np.linspace(-2, 2, 100)
b = np.linspace(-2, 2, 100)
A, B = np.meshgrid(a, b)
L = np.cos(A**2.4) + np.sin(B**2.4)

fig = go.Figure(data = [go.Surface(x = a, y = b, z = L, opacity = 0.5)]) # colorscale = "viridis" - cmap처럼 색상 변경
fig.update_traces(contours_z = dict(show = True, usecolormap = True, highlightcolor = "limegreen", project_z = True)) # project_z 컨투어
fig.update_layout(title = "cos(A^2.4) + sin(B^2.4)", width = 700, height = 600,
                  scene = dict(zaxis = dict(nticks = 40, range = [-3, 3])))
```

이미지 보기

```python
import matplotlib.pyplot as plt
from torchvision import datasets
test_DS = datasets.MNIST(root="data/", train = False, download = True)
```

```python
print(test_DS.data.shape) # (개, 행, 렬)
print(test_DS.data[0])
```

```python
plt.imshow(test_DS.data[0]) # image show
```

```python
plt.imshow(test_DS.data[0], cmap = "gray") # cmap = "gray"할 때 흑백만 흑백으로 출력됨. 컬러는 해당없음
plt.colorbar() # 우측 컬러바 출력
```

```python
plt.figure(figsize = (7, 4))
for i in range(6):
  plt.subplot(2, 3, i+1, xticks = [], yticks = []) # xticks = [], yticks = [] 격자 수치 삭제
  plt.imshow(test_DS.data[i], cmap = "gray")
  plt.title(test_DS.targets[i].item(), color = "m") # .item 사용해서 "tensor()" 삭제
```

```python
im = test_DS.data[0]
plt.figure(figsize = (20, 15))
plt.imshow(im, cmap = "gray")
plt.colorbar()

for i in range(im.shape[0]):
  for j in range(im.shape[1]):
    plt.text(j, i, im[i, j].item(), # (x위치, y위치, 쓰고싶은 글자)
             horizontalalignment = "center", # 글씨 밀림방지 가운데 정렬
             color = "k" if im[i, j] > torch.max(im)/1.5 else "w")


plt.xticks([])
plt.yticks([])
```
